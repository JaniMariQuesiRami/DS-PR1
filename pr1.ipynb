#### Combinar todos los departamentos en un solo csv y df

```python
import os
import pandas as pd
from bs4 import BeautifulSoup

def read_html_with_encoding(file_path, encodings=['utf-8', 'ISO-8859-1']):
    for encoding in encodings:
        try:
            with open(file_path, 'r', encoding=encoding) as file:
                return file.read()
        except UnicodeDecodeError:
            continue
    raise ValueError(f"No se pudo leer el archivo {file_path} con las codificaciones proporcionadas.")

# Especifica el directorio donde est√°n los archivos HTML
input_dir = './data'
output_dir = './data'

# Lista todos los archivos HTML en el directorio
html_files = [f for f in os.listdir(input_dir) if f.endswith('.html')]

for html_file in html_files:
    # Construye la ruta completa del archivo HTML
    html_path = os.path.join(input_dir, html_file)
    
    # Lee el contenido del archivo HTML con la codificaci√≥n adecuada
    html_content = read_html_with_encoding(html_path)

    # Usa BeautifulSoup para analizar el contenido HTML
    soup = BeautifulSoup(html_content, 'html.parser')

    # Encuentra la tabla en el HTML
    table = soup.find('table', {'id': '_ctl0_ContentPlaceHolder1_dgResultado'})

    # Verifica si la tabla se encontr√≥
    if table:
        # Usa pandas para leer la tabla en un DataFrame
        df = pd.read_html(str(table))[0]

        # Construye el nombre del archivo CSV
        csv_file = html_file.replace('.html', '.csv')
        csv_path = os.path.join(output_dir, csv_file)

        # Guarda el DataFrame en un archivo CSV
        df.to_csv(csv_path, index=False, encoding='utf-8')
        print(f'Archivo CSV guardado: {csv_path}')
    else:
        print(f'No se encontr√≥ la tabla en el archivo HTML: {html_file}')
```

```python
# Especifica el directorio donde est√°n los archivos CSV
path = './data'

# Utiliza glob para encontrar todos los archivos CSV en el directorio
all_files = glob.glob(os.path.join(path, "*.csv"))

# Lista para almacenar los DataFrames
df_list = []

# Itera sobre cada archivo y lee el CSV en un DataFrame
for filename in all_files:
    df = pd.read_csv(filename)
    df_list.append(df)

# Concatena todos los DataFrames en uno solo
combined_df = pd.concat(df_list, ignore_index=True)

# save into a csv file
combined_df.to_csv('combined_df.csv', index=False)
```

#### Se realizaron las siguientes acciones de limpieza en los datos:

- PLAN: Se corrigieron errores en la carga de texto donde las tildes se mostraban incorrectamente. Espec√≠ficamente, se reemplazaron las entradas "DOS DÕÅS A LA SEMANA" por "DOS D√çAS A LA SEMANA" y "UN DÕÅ A LA SEMANA" por "UN D√çA A LA SEMANA".

- JORNADA, MODALIDAD, STATUS, AREA, SECTOR: Los valores faltantes (NaN) en estas columnas se reemplazaron por "NO ESPECIFICADA" para evitar datos incompletos.

- NIVEL: Se filtraron los datos para eliminar todos los registros cuyo valor en la columna 'NIVEL' no sea "DIVERSIFICADO", manteniendo √∫nicamente los registros que cumplen con este criterio.

```python
# 1. Corregir las palabras con tilde mal cargadas en la columna 'PLAN'
combined_df['PLAN'] = combined_df['PLAN'].replace({
    'SEMIPRESENCIAL (DOS DÕÅS A LA SEMANA)': 'SEMIPRESENCIAL (DOS D√çAS A LA SEMANA)',
    'SEMIPRESENCIAL (UN DÕÅ A LA SEMANA)': 'SEMIPRESENCIAL (UN D√çA A LA SEMANA)'
})

# 2. Cambiar los valores NaN por 'NO ESPECIFICADA' en las columnas 'JORNADA', 'MODALIDAD', 'STATUS', 'AREA', y 'SECTOR'
columns_to_fill = ['JORNADA', 'MODALIDAD', 'STATUS', 'AREA', 'SECTOR']

combined_df[columns_to_fill] = combined_df[columns_to_fill].fillna('NO ESPECIFICADA')

# 3. Remover todos los registros cuyo 'NIVEL' no sea 'DIVERSIFICADO'
combined_df = combined_df[combined_df['NIVEL'] == 'DIVERSIFICADO']


```
